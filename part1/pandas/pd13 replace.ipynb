## pd13. 값 변경

```python
import pandas as pd

df = pd.read_csv('data/cafe3.csv')
df.head(3)
```

### 1. replace() 활용

```python
df.replace('아메리카노', '룽고', inplace=True)
df.replace('녹차','그린티',inplace=True)
df
```

```python
change = {'룽고':'아메리카노', '그린티':'녹차'}
df.replace(change,inplace=True)
df
```

### 2. loc/iloc 활용

```python
df.loc[6,'원산지'] ='대한민국'
df.tail(3)
```

```python
df.loc[1:2,'이벤트가'] = 1000
df.head()
```

```python
df.to_csv("data/cafe4.csv", index=False)
```
