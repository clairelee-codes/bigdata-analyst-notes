## pd10. 정렬

```python
import pandas as pd

df = pd.read_csv('data/cafe2.csv')
df.head(2)
```

### 1. 정렬 방법
- 인덱스 기준: sort_index()
- 데이터 값 기준: sort_values()

### 2. 인덱스 기준 정렬

```python
df.sort_index(ascending=False)
```

### 3. 데이터 값 기준 정렬

```python
df.sort_values('메뉴',ascending=False)
```

### 4. 2개 이상의 기준 정렬

```python
df.sort_values(['가격','메뉴'],ascending=[False,True],inplace=True)
df
```

### 5. 인덱스 초기화

```python
df.reset_index(drop=True)
```
